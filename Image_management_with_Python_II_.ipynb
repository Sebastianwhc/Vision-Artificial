And what does this image look like? How can we visualize it?

```python
# To display images
import matplotlib.pyplot as plt
```

```python
# Define a function to display an image on the screen with the criteria that I consider most accurate
def imshow(img):
      fig,ax=plt.subplots(figsize=(7,7))

      # The command that actually displays the image
      ax.imshow(img,cmap=plt.cm.gray)

    # To prevent numbers from appearing on the axes
      ax.set_xticks([]),ax.set_yticks([])
      plt.show()
```

```python
 imshow(coins)
```

It seems then that if we want to work with images, we also need the *numpy* package which is one of the most common and used in data processing.

```python
import numpy as np
```

The reason we run it as `as np` is to facilitate later reference to said package. In this way, and by typing *np.* plus the tab key, we can see all the options and functions that this package offers us.

Let's now calculate the maximum of the matrix. As we have seen before, it is a grayscale image. Then, if there is a very white pixel (represented by 255), this will be the maximum.

```python
np.max(coins)
```

Therefore, the maximum of this image is 252, which is very close to the total maximum (255).

On the other hand, the minimum would be calculated similarly with `np.min(coins)`.

And how would we do it to calculate the maximum per row and column? Very simple, for that, we have to use the *axis* parameter which indicates that the specific *numpy* operation that we want to execute will be carried out by columns *axis=0* or by rows *axis=1*.

Let's look at it first with a simpler example, before going to the image.

```python
# Tomemos un array de 16 elementos y hagamos un array de 4x4
a=np.arange(16).reshape((4,4))
```

Let's calculate the maximum as before:

```python
a
```

Now we are going to calculate the maximum by columns, using the parameter *axis=0*. In general, to see all the parameter alternatives that a function offers us, we must press tab after a comma.

```python
np.max(a)
```

```python
# Máximos por columnas
np.max(a, axis=0)
```

```python
# Máximos por filas
np.max(a, axis=0)
```

And returning to our example in the image, we are going to calculate the maximum by columns and by rows.

```python
# Máximos por columnas
max_per_colum=np.max(coins,axis=0)
max_per_colum
```

```python
# Máximos por filas
max_per_row=np.max(coins,axis=1)
max_per_row
```

```python
# Let's paint the image again, see if we see something
plt.plot(max_per_row)
plt.show()
```

Let's binarize the previous signal. That is, we are going to set a threshold and everything that is greater than that threshold we will set to a value of 255 (white) and everything that is below, we will set to a value of 0 (black).

```python
umbral = 160
binary_max_per_row = 255*(max_per_row>umbral)
```

```python
plt.plot(binary_max_per_row)
plt.show
```

```python
# Hacemos una copia de la imagen para hacer nuestras pruebas
test_coins=coins.copy()
```

```python
# Todos aquellas posiciones del vector que tengan valores que iguales a 0 los modifico en la imagen poniéndolos
# también a 255.
test_coins[binary_max_per_row==0,:]=255
```

```python
imshow(test_coins)
```

This way, I'm left with an estimate of where the coins are. These types of strategies serve to detect aligned objects (such as the information within a DNI).

If we repeat the same operation for the vertical components, let's see what happens.

```python
# Máximos por columnas
mas_per_colums=np.max(coins,axis=0)
```

```python
# Vamos a representar ahora el máximo por filas
plt.plot(mas_per_colums)
plt.show()
```

Aquí el umbral es más complicado de elegir. En este caso, podemos ver que existen 6 mínimos, correspondientes a las 6 monedas que tenemos por cada fila.

Se puede decir que estos mínimos están, aproximadamente (que es lo que buscamos ahora), en las posiciones: 75, 130, 180, 240, 290.

Por supuesto, estos mínimos se pueden obtener de forma más elegante con un algoritmo de detección de mínimos locales (from scipy.signal import argrelextrema)

```python
umbral = 160
binary_max_per_colum = 255*(max_per_colum>umbral)
```

```python
plt.plot(binary_max_per_colum)
plt.show
```
